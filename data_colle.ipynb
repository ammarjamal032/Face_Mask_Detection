```python
import numpy as np
import cv2
import os
import sys

def with_mask():
    # Load the cascade
    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
    # To capture video from webcam.
   
    cap = cv2.VideoCapture('Malala Yousafzai - The right to learning should be given to any child.mp4')
    
    # To use a video file as input
    # cap = cv2.VideoCapture('filename.mp4')
    count=0
    while True:
        # Read the frame
        _, img = cap.read()
        # Convert to grayscale
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        # Detect the faces
        faces = face_cascade.detectMultiScale(gray, 1.3, 5)#1.3, 5
        # Draw the rectangle around each face

        for (x, y, w, h) in faces:
            count = count + 1
            #name = './images/0/' + str(count) + '.jpg'   #with mask
            name = './images/1/' + str(count) + '.jpg'  #without mask
            print('Creating Images............' + name)
            cv2.imwrite(name, img[y:y + h, x:x + w])
            img = cv2.rectangle(img, (x, y), (x + w, y + h), (255, 255, 0), 10)
            roi_gray = gray[y:y + h, x:x + w]
            roi_color = img[y:y + h, x:x + w]
            eyes = eye_cascade.detectMultiScale(roi_gray)
            for (ex, ey, ew, eh) in eyes:
                cv2.rectangle(roi_color, (ex, ey), (ex + ew, ey + eh), (0, 255, 0), 5)
        # Display
        cv2.imshow('img', img)
        # Stop if escape key is pressed
        cv2.waitKey(30) & 0xff
        if count > 4000:
            break
    # Release the VideoCapture object
    cap.release()
    cv2.destroyAllWindows()
    
    
def without_mask():
    # Load the cascade
    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
    # To capture video from webcam.
    
    cap = cv2.VideoCapture(0)
    
    # To use a video file as input
    # cap = cv2.VideoCapture('filename.mp4')
    count=0
    while True:
        # Read the frame
        _, img = cap.read()
        # Convert to grayscale
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        # Detect the faces
        faces = face_cascade.detectMultiScale(gray, 1.3, 5)#1.3,5
        # Draw the rectangle around each face

        for (x, y, w, h) in faces:
            count = count + 1
            #name = './images/0/' + str(count) + '.jpg'   #with mask
            #name = './images/1/' + str(count) + '.jpg'  #without mask
            print('Creating Images............' + name)
            cv2.imwrite(name, img[y:y + h, x:x + w])
            img = cv2.rectangle(img, (x, y), (x + w, y + h), (255, 255, 0), 10)
            roi_gray = gray[y:y + h, x:x + w]
            roi_color = img[y:y + h, x:x + w]
            eyes = eye_cascade.detectMultiScale(roi_gray)
            for (ex, ey, ew, eh) in eyes:
                cv2.rectangle(roi_color, (ex, ey), (ex + ew, ey + eh), (0, 255, 0), 5)
        # Display
        cv2.imshow('img', img)
        # Stop if escape key is pressed
        cv2.waitKey(30) & 0xff
        if count > 4000:
            break
    # Release the VideoCapture object
    cap.release()
    cv2.destroyAllWindows()
```
